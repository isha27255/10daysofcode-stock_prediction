```python
import pandas as pd
import numpy as np
# !pip install matplotlib
from matplotlib import pyplot as plt 
```

```python
# Reading csv and creating a dataframe
df = pd.read_csv("housing.csv")
df
# 'RM' is the average number of rooms among homes in the neighborhood.
# 'LSTAT' is the percentage of homeowners in the neighborhood considered "lower class" (working poor).
# 'PTRATIO' is the ratio of students to teachers in primary and secondary schools in the neighborhood.
# 'MEDV' is  the final price
```

```python
# dropping columns
df2 = df.drop(["PTRATIO"], axis=1)
df2
```

```python
# Rename a column
df2.rename(columns = {'LSTAT':'lower class HO'}, inplace = True) 
df2
```

```python
# converting to 2D array 
dataset = df.values
dataset
```

```python

# plot any two columns
X = dataset[:,0].reshape(-1,1)
Y = dataset[:,-1]

plt.plot(X,Y, '.')
plt.xlabel("RM")
plt.ylabel("MEDV")
plt.show()
```

```python
# plot any two columns
X = dataset[:,1].reshape(-1,1)
Y = dataset[:,-1]

plt.plot(X,Y, '.')
plt.xlabel("LSTAT")
plt.ylabel("MEDV")
plt.show()
```

```python
# plot any two columns
X = dataset[:,2]
Y = dataset[:,-1]

plt.plot(X,Y, '.')
plt.xlabel("PTRATIO")
plt.ylabel("MEDV")
plt.show()
```
